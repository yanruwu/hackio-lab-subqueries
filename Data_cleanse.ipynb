```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
df_places = pd.read_csv("datos/api_foursquare.csv", index_col = 0)
df_aemet = pd.read_csv("datos/df_aemet_final.csv", index_col = 0)
df_municipios = pd.read_csv("datos/municipios.csv", index_col = 0).T

df_places.set_index("municipio", inplace=True)
df_places.index.name = "municipio_id"

df_aemet.reset_index(inplace=True)
df_splitted = df_aemet["localizacion_id"].str.split("-id", expand=True)
df_aemet[["municipio_id", "code"]] = df_splitted

df_aemet.drop(columns="localizacion_id", inplace=True)
df_aemet.set_index("municipio_id", inplace=True)

df_municipios.rename(columns={0 :"latitude", 1 :"longitude", 2 :"nombre"}, inplace=True)
df_municipios.index.name = "municipio_id"
```

```python
df_aemet["fecha"]=df_aemet["fecha"].map(pd.to_datetime)
```

```python
df_aemet["hora"] = df_aemet["fecha"].apply(lambda d: d.time())
df_aemet["fecha"] = df_aemet["fecha"].apply(lambda d: d.date())
```

```python
df_aemet[df_aemet.columns[df_aemet.columns.str.contains("prob")]] = df_aemet[df_aemet.columns[df_aemet.columns.str.contains("prob")]].astype(object)
```

```python
df_aemet["prob._precip.__(%)"][df_aemet["prob._precip.__(%)"] == "Riesgo"] = "100"
df_aemet["prob._precip.__(%)"] = pd.cut(df_aemet["prob._precip.__(%)"].astype(int), bins=[-0.1,25,50,75,100], labels = ["Baja", "Media", "Alta", "Muy alta"])
df_aemet["prob._de_nieve_(%)"] = pd.cut(df_aemet["prob._de_nieve_(%)"].astype(int), bins=[-0.1,25,50,75,100], labels = ["Baja", "Media", "Alta", "Muy alta"])
df_aemet["prob._de_tormenta_(%)"] = pd.cut(df_aemet["prob._de_tormenta_(%)"].astype(int), bins=[-0.1,25,50,75,100], labels = ["Baja", "Media", "Alta", "Muy alta"])
```

```python
def second_stamp(time_of_day):
    seconds = time_of_day.hour*3600 + time_of_day.minute*60 + time_of_day.second
    return seconds
df_aemet["daystage"] = pd.qcut(df_aemet["hora"].apply(second_stamp), q = 4, labels= ["madrugada", "mediodia", "tarde", "noche"])
```

```python
df_aemet.fecha = df_aemet.fecha.astype('datetime64[s]')
```

```python
df_aemet.drop(columns=["avisos", "hora", "code"], inplace=True)
```

```python
df_aemet_grouped = df_aemet.groupby(["municipio_id", "fecha", "daystage"], observed=True)["cielo prob._precip.__(%)	prob._de_nieve_(%)	prob._de_tormenta_(%)	dirección_viento".split() ].agg(pd.Series.mode).reset_index()
```

```python
from geopy.geocoders import OpenCage
import dotenv
import os

dotenv.load_dotenv()
token = os.getenv("token")
geolocator = OpenCage(api_key=token)
```

```python
defect = []
for muni in list(df_aemet.index):
    if muni not in df_municipios.index :
        defect.append(muni)

defect_list = np.unique(defect)
```

```python
def get_coords(location_list, token):
    """
    Obtiene las coordenadas geográficas (latitud y longitud) de una lista de ubicaciones.

    Parámetros:
    ----------
    location_list : list
        Lista de ubicaciones en formato de texto (por ejemplo, nombres de ciudades o direcciones).

    Retorno:
    -------
    pandas.DataFrame
        Un DataFrame con las columnas "Nombre", "Latitud" y "Longitud", donde "Nombre" es el nombre normalizado
        de la ubicación, y "Latitud" y "Longitud" corresponden a las coordenadas geográficas de cada ubicación.
    """
    loc_dict = dict()
    names = []
    lat = []
    long = []
    municipio_id = []

    geolocator = OpenCage(api_key=token)
    
    for loc in (location_list):
        municipio_id.append(loc)
        loc_get = geolocator.geocode(loc)
        names.append(loc_get.raw["components"]["_normalized_city"])
        lat.append(loc_get.latitude)
        long.append(loc_get.longitude)
    loc_dict["municipio_id"] = municipio_id
    loc_dict["latitude"] = lat
    loc_dict["longitude"] = long
    loc_dict["nombre"] = names

    return pd.DataFrame(loc_dict)
```

```python
dict_defect = dict(municipio_id = [], latitude = [], longitude = [], nombre = [])
# for m in defect_list:
df_defect = get_coords(defect_list, token= token)

df_defect = df_defect.set_index("municipio_id")
```

```python
df_municipios = pd.concat([df_municipios, df_defect])
df_municipios.head()
```
